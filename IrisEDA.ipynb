# (ANOVA) Comparing Mean Petal Lengths 

```python
#step 1
# Overall mean
overall_mean = iris['PetalLengthCm'].mean()

# Group statistics
group_means = iris.groupby('Species')['PetalLengthCm'].mean()
group_sizes = iris.groupby('Species')['PetalLengthCm'].count()

# Total Sum of Squares (SSTotal)
ss_total = np.sum((iris['PetalLengthCm'] - overall_mean) ** 2)

# Between-Group Sum of Squares (SSBetween)
ss_between = np.sum(group_sizes * (group_means - overall_mean) ** 2)

# Within-Group Sum of Squares (SSError)
ss_within = np.sum(iris.groupby('Species').apply(
    lambda group: np.sum((group['PetalLengthCm'] - group['PetalLengthCm'].mean()) ** 2)
))

print(f"SS Total: {ss_total:.2f}")
print(f"SS Between (Explained): {ss_between:.2f}")
print(f"SS Within (Error): {ss_within:.2f}")
```

```python
#step 2 
# Number of groups
k = iris['Species'].nunique()  # 3 species

# Total number of observations
N = len(iris)  # 150 observations

# Degrees of freedom
df_between = k - 1
df_within = N - k
df_total = N - 1

print(f"Between-Group Degrees of Freedom (df_between): {df_between}")
print(f"Within-Group Degrees of Freedom (df_within): {df_within}")
print(f"Total Degrees of Freedom (df_total): {df_total}")
```

```python
#step 3 
# Degrees of Freedom
k = iris['Species'].nunique()  # Number of groups (3 species)
N = len(iris)  # Total number of observations (150)

df_between = k - 1  # Degrees of freedom for treatment
df_within = N - k   # Degrees of freedom for error

# Mean Square Treatment (MS Treatment)
ms_treatment = ss_between / df_between

# Mean Square Error (MS Error)
ms_error = ss_within / df_within

print(f"MS Treatment: {ms_treatment:.2f}")
print(f"MS Error: {ms_error:.2f}")
```

```python
#step 8 
f_stat = ms_treatment / ms_error
print(f"F-Statistic: {f_stat:.2f}")
```

```python
# same as the steps 1 , 2 ,3 
# function f_oneway  do the job  to get the F-statistic and from the table we get the p value 
from scipy.stats import ttest_ind, f_oneway, chi2_contingency

anova_result = f_oneway(iris_Setosa, iris_Versicolor, iris_Virginca)
print(f"F-statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")
```

```python
#Fo is F-statistic and 1179.0343277002194 
#and from Fo we het P value 
#compare the P value with alpha 0.05 for the error 
# if p value  > alpha meaning fail to reject h1 is true 
# if p value < alpha meaning reject h0 
# 3.0519758018278374e equal to 0  and 0 < 0.05 that mean h0 is rejected and At least one species has a different mean petal length
```

# (Chi-Square Test) Independence Between Species and Petal Length Categories 

```python
# Categorize petal length by using binning for each flower 
# Conver the pl to Category 
iris['petal_length_category'] = pd.cut(
    iris['PetalLengthCm'],
    bins=[0, 2, 4, 6, 8],
    labels=['Short', 'Medium', 'Long', 'Very Long']
)
```

```python
# Create a contingency table
# see the distrubution for the pl based in the flower 
contingency_table = pd.crosstab(iris['Species'], iris['petal_length_category'])
contingency_table
```

```python
# Perform Chi-Square test
# get the expected frequency and p value for it 
# note alpha is 0.5 
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat}, p-value: {p_val}")
```

```python
#expected frequency is Chi-Square Statistic  188.48
#and from expected frequency we get X^2 and from it we get p value is 5.354817569315289e-38
#compare the P value with alpha 0.05  
# if p value  > alpha meaning fail to reject h1 is true 
# if p value < alpha meaning reject h0 
    # 5.354817569315289e-38 equal to 0  and 0 < 0.05 that mean h0 is rejected and indicates that species and petal length categories are not independent
    # There is an extremely strong association between the two variables being tested species and petal length categories (depend on each other )
```
